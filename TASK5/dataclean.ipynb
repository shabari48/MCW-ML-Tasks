```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('dataset/bank/bank-additional-full.csv',sep=';')
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
important_features = [
        'age',          # Client demographics
        'job',
        'education',
        'default',      # Financial status
        'housing',
        'loan',
        'duration',     # Call/Campaign info
        'campaign',
        'previous',
        'poutcome',
        'emp.var.rate', 
        'cons.price.idx',
        'cons.conf.idx',
        'euribor3m',
        'nr.employed',
        'y'            
    ]
```

```python
df=data.loc[:,important_features].copy()
df.head()
```

```python
# Rename columns
df.columns = [
    'Age', 'Job', 'Education', 'Default', 'Housing', 'Loan', 
    'Duration', 'Campaign', 'Previous', 'Poutcome', 'EmpVarRate', 
    'ConsPriceIdx', 'ConsConfIdx', 'Euribor3m', 'NrEmployed', 'Y'
]

# Convert categorical columns to numerical
categorical_columns = ['Job', 'Education', 'Default', 'Housing', 'Loan', 'Poutcome', 'Y']
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True,dtype=int)

df_encoded.head(20)
```
